```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Preparation

```python
# Read the CSV file located at "ITCS6100_training_dataset/training_data.csv"
# and create a pandas DataFrame from the data in the CSV file.
# Store the DataFrame in the variable 'df'
df = pd.read_csv("ITCS6100_training_dataset/training_data.csv")
df
```

### Data Cleaning 

```python
# Section 1: Handling Missing Values

## Step 1: Count Missing Values in the DataFrame
missing_values_count = df.isnull().sum()

## Step 2: Filter and Count Attributes with Missing Values
n_missing_attributes = (missing_values_count > 0).sum()

## Step 3: Print Number of Attributes with Missing Values
print(f"Number of attributes with missing values: {n_missing_attributes}")

## Step 4: Print Count of Missing Values for Each Attribute (if any)
if n_missing_attributes > 0:
    print("Count of missing values for each attribute:")
    print(missing_values_count[missing_values_count > 0])
```

```python
## Step 5: Analyze and Handle Specific Columns with Missing Values

### Vehicle_Bodily_Injury_Limit Column
# Calculate the count of missing values in the 'Vehicle_Bodily_Injury_Limit' column
vehicle_missing_count = df['Vehicle_Bodily_Injury_Limit'].isnull().sum()

# Calculate the percentage of missing values in the 'Vehicle_Bodily_Injury_Limit' column
vehicle_missing_percentage = vehicle_missing_count / len(df) * 100

# Print the percentage of missing values in 'Vehicle_Bodily_Injury_Limit' column
print(f"Percentage of missing values in 'Vehicle_Bodily_Injury_Limit': {vehicle_missing_percentage}%")

# Remove extra spaces in the 'Vehicle_Bodily_Injury_Limit' column values
df['Vehicle_Bodily_Injury_Limit'] = df['Vehicle_Bodily_Injury_Limit'].str.strip()

# Find the mode of the 'Vehicle_Bodily_Injury_Limit' column
mode_vehicle_limit = df['Vehicle_Bodily_Injury_Limit'].value_counts().idxmax()

# Fill missing values in 'Vehicle_Bodily_Injury_Limit' column with the mode
df['Vehicle_Bodily_Injury_Limit'] = df['Vehicle_Bodily_Injury_Limit'].fillna(mode_vehicle_limit)

# Print unique values in 'Vehicle_Bodily_Injury_Limit' column after cleaning
print(f"Unique values in 'Vehicle_Bodily_Injury_Limit' after cleaning: {df['Vehicle_Bodily_Injury_Limit'].unique()}")

# Count the number of missing values in the 'EEA_Prior_Bodily_Injury_Limit' column
ee_missing_count = df['EEA_Prior_Bodily_Injury_Limit'].isnull().sum()

# Calculate the percentage of missing values in the 'EEA_Prior_Bodily_Injury_Limit' column
ee_missing_percentage = ee_missing_count / len(df) * 100

# Print the percentage of missing values in 'EEA_Prior_Bodily_Injury_Limit' column
print(f"Percentage of missing values in 'EEA_Prior_Bodily_Injury_Limit': {ee_missing_percentage}%")

# Remove extra spaces in the 'EEA_Prior_Bodily_Injury_Limit' column values
df['EEA_Prior_Bodily_Injury_Limit'] = df['EEA_Prior_Bodily_Injury_Limit'].str.strip()

# Find the mode of the 'EEA_Prior_Bodily_Injury_Limit' column
mode_ee_limit = df['EEA_Prior_Bodily_Injury_Limit'].value_counts().idxmax()

# Fill missing values in 'EEA_Prior_Bodily_Injury_Limit' column with the mode
df['EEA_Prior_Bodily_Injury_Limit'] = df['EEA_Prior_Bodily_Injury_Limit'].fillna(mode_ee_limit)

# Print unique values in 'EEA_Prior_Bodily_Injury_Limit' column after cleaning
print(f"Unique values in 'EEA_Prior_Bodily_Injury_Limit' after cleaning: {df['EEA_Prior_Bodily_Injury_Limit'].unique()}")
```

```python
# Section 2: Checking for Duplicates

## Step 6: Check for Duplicated Rows
duplicates_count = df.duplicated().sum()

## Step 7: Print Number of Duplicated Rows
if duplicates_count == 0:
    print("There are no duplicates in the DataFrame.")
else:
    print(f"Number of duplicated rows: {duplicates_count}")
```

### Explorative Data Analysis

```python
# Section 3: Filtering Numeric Columns

## Step 8: Filter Numeric Columns in the DataFrame
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

## Step 9: Output Numeric Column Names
print("Numeric columns in the DataFrame:")
print(numeric_columns.columns)
```

```python
## Step 10: Calculate the correlation matrix for numeric columns
correlation_matrix = numeric_columns.corr()

## Step 11: Plot the heatmap of the correlation matrix
f, ax = plt.subplots(figsize=(12, 11))
plt.title('Heatmap of Correlation matrix', fontsize=20)
sns.heatmap(correlation_matrix, ax=ax, cmap="coolwarm", linewidths=0.1,cbar_kws={"extend": "both", "extendfrac": 0.05, "extendrect": True})

## Step 12: Display the heatmap
plt.show()
```
